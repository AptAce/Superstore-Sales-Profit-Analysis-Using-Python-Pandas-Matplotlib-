```python
# 7. Which sub-category has the lowest total sales?

subcat_sales = Df.groupby('Sub.Category')['Sales'].sum()
print(subcat_sales)
```

```python
lowest_sales = subcat_sales.min()
print(lowest_sales)
```

```python
lowest_subcat = subcat_sales.idxmin()
lowest_sales = subcat_sales.min()

print("Sub-category with the lowest total sales:", lowest_subcat)
print("Total sales for this sub-category: ${:,.2f}".format(lowest_sales))
```

```python
# Df.groupby('Sub-Category')['Sales'].sum() → Groups by sub-category and sums sales.

# .idxmin() → Finds the sub-category with the smallest total sales.

# .min() → Gives the sales value of that sub-category.

# This tells you exactly which sub-category is underperforming in terms of sales.
```

```python
# 8. What is the average sales value per order?

average_sales = Df['Sales'].mean()
print("Average sales per order: ${:,.2f}".format(average_sales))
```

```python
# Df['Sales'] → Selects the Sales column.

# .mean() → Calculates the average of all sales values.

# "{:,.2f}".format() → Formats it nicely as currency with two decimal points and commas.

# This tells you on average how much each order contributes to total sales.
```

```python
# 10. What is the total profit made?

total_profit = Df['Profit'].sum()
print("Total profit made: ${:,.2f}".format(total_profit))
```

```python
# 11. Which region generates the highest profit?
region_profit = Df.groupby('Region')['Profit'].sum()

top_region = region_profit.idxmax()
highest_profit = region_profit.max()

print("Region with the highest profit:", top_region)
print("Total profit for this region: ${:,.2f}".format(highest_profit))
```

```python
# 12. Which category is the most profitable?

category_profit = Df.groupby('Category')['Profit'].sum()

top_category = category_profit.idxmax()
highest_profit = category_profit.max()

print("Most profitable category:", top_category)
print("Total profit for this category: ${:,.2f}".format(highest_profit))
```

```python
# 13. Which sub-category has the lowest average profit?

subcat_avg_profit = Df.groupby('Sub.Category')['Profit'].mean()

lowest_subcat = subcat_avg_profit.idxmin()
lowest_avg_profit = subcat_avg_profit.min()

print("Sub-category with the lowest average profit:", lowest_subcat)
print("Average profit for this sub-category: ${:,.2f}".format(lowest_avg_profit))
```

```python
# 14. What is the total quantity of items sold?

total_quantity = Df['Quantity'].sum()

print("Total quantity of items sold:", total_quantity)
```

```python
# 15. Which category sells the highest quantity of products?

category_quantity = Df.groupby('Category')['Quantity'].sum()

# Category with highest total quantity
top_category = category_quantity.idxmax()
highest_quantity = category_quantity.max()

print("Category that sells the highest quantity:", top_category)
print("Total quantity sold in this category:", highest_quantity)
```

```python
# 16. What is the average discount given across all orders?

average_discount = Df['Discount'].mean()

print("Average discount across all orders:", round(average_discount, 4))
```

```python
# 17. Which region gives the highest average discount?

region_avg_discount = Df.groupby('Region')['Discount'].mean()

top_region = region_avg_discount.idxmax()
highest_avg_discount = region_avg_discount.max()

print("Region with the highest average discount:", top_region)
print("Highest average discount: {:.2%}".format(highest_avg_discount))
```

```python
# 18. How does total sales change over time?

Df['Order.Date'] = pd.to_datetime(Df['Order.Date'])
sales_by_year = Df.groupby(Df['Order.Date'].dt.year)['Sales'].sum()

print("Total Sales Per Year:")
print(sales_by_year)
```

```python
import matplotlib.pyplot as plt

sales_by_year.plot(kind='line', marker='o')

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
# 19. Which month records the highest sales?

Df['Month'] = Df['Order.Date'].dt.month_name()
monthly_sales = Df.groupby('Month')['Sales'].sum()
top_month = monthly_sales.idxmax()
highest_sales = monthly_sales.max()

print("Month with the highest sales:", top_month)
print("Total sales in that month: ${:,.2f}".format(highest_sales))
```
